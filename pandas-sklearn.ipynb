```python
# Verify the duplicated data
df = pandas.DataFrame(dataList)
print("Duplicate Rows :")
df[df.duplicated()]
```

```python
# Remove useless features
df = df.drop("Name", axis = 1)
df = df.drop("LoanNr_ChkDgt", axis = 1)
```

```python
# Deal with the null values - ChgOffDate 
import numpy as np

df["ChgOffDate"].fillna(method="bfill")
```

```python
# Deal with the null values - MIS_Status 
import numpy as np

df = df.dropna(subset=["MIS_Status"])
print(df.shape)
```

```python
# Deal with other null values
df = df.dropna(subset=["City","State", "Bank", "BankState", "NewExist", "RevLineCr", "LowDoc", "DisbursementDate"])
print(df.shape)
```

```python
# Break the dataset 

middle_index = len(df)// 2
print(middle_index)

first_dataset = df[:middle_index]
second_dataset = df[middle_index:]

print(first_dataset.head())
print(second_dataset.head())
```

```python
# Merge the first and second datasets again
df = first_dataset.merge(second_dataset, how="outer")
print(len(df))
print(df.head())
```

```python
# 6.1 Transform data
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

df["City"] = le.fit_transform(df["City"])
df["State"] = le.fit_transform(df["State"])
df["RevLineCr"] = le.fit_transform(df["RevLineCr"])
df["Bank"] = le.fit_transform(df["Bank"])
df["BankState"] = le.fit_transform(df["BankState"])
df["NAICS"] = le.fit_transform(df["NAICS"])
df["ApprovalDate"] = le.fit_transform(df["ApprovalDate"])
df["LowDoc"] = le.fit_transform(df["LowDoc"])
df["ChgOffDate"] = le.fit_transform(df["ChgOffDate"])
df["DisbursementGross"] = le.fit_transform(df["DisbursementGross"])
df["DisbursementDate"] = le.fit_transform(df["DisbursementDate"])
df["BalanceGross"] = le.fit_transform(df["BalanceGross"])
df["GrAppv"] = le.fit_transform(df["GrAppv"])
df["SBA_Appv"] = le.fit_transform(df["SBA_Appv"])
df["ChgOffPrinGr"] = le.fit_transform(df["ChgOffPrinGr"])
df["MIS_Status"] = le.fit_transform(df["MIS_Status"])
```

```python
# 6.2 Get the distributions
import seaborn as sns
sns.set(rc={'figure.figsize':(16.7,13.27)})

sns.violinplot(x=df["MIS_Status"])
```

```python
# 6.3 balance datasets
import numpy as np

target_count1, target_count0 = df["MIS_Status"].value_counts()

MIS_Status_0 = df[df["MIS_Status"] == 0]
MIS_Status_1 = df[df["MIS_Status"] == 1]

MIS_Status_1_under = MIS_Status_1.sample(target_count0)
df = pandas.concat([MIS_Status_0, MIS_Status_1_under], axis=0)

print(df["MIS_Status"].value_counts())
sns.violinplot(x=df["MIS_Status"])
```

```python
# Data Mining Step1: spliting dataset into testing and training datasets

import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn import metrics

# feature_cols = ['City', 'State', 'Zip', 'Bank', 'BankState', 'NAICS', 
#                 'ApprovalDate','Term', 'NoEmp', 'NewExist', 
#                 'CreateJob', 'RetainedJob','FranchiseCode', 'UrbanRural', 
#                 'RevLineCr', 'LowDoc', 'ChgOffDate','DisbursementDate', 
#                 'DisbursementGross', 'BalanceGross','ChgOffPrinGr', 
#                 'GrAppv', 'SBA_Appv']

feature_cols = ['City', 'State', 'Zip', 'Bank', 'BankState', 'NAICS', 
                'ApprovalDate','Term', 'NoEmp', 'NewExist', 'CreateJob', 'RetainedJob',
                'FranchiseCode', 'UrbanRural', 'RevLineCr', 'LowDoc', 'DisbursementGross', 
                'BalanceGross', 'GrAppv', 'SBA_Appv']

X = df[feature_cols]

# X = df.drop(columns = "MIS_Status")
y = df["MIS_Status"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) 
```

```python
# Data Mining Model: Decision Tree training data
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model = model.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = model.predict(X_test)

# Features importance
features_df = X.columns
importances_df = model.feature_importances_
feature_importances_df = pd.DataFrame()
feature_importances_df['featureName'] = features_df
feature_importances_df['featureImportance'] = importances_df
feature_importances_df = feature_importances_df.sort_values('featureImportance', ascending= False)

print(feature_importances_df)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
